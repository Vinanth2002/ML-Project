### BIVARIATE ANALYSIS ( Is lunch type intake has any impact on student's performance ? )


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# If not already created
df['total_score'] = df[['math_score', 'reading_score', 'writing_score']].sum(axis=1)
df['average_score'] = df['total_score'] / 3

# Group by lunch and calculate mean of each score
lunch_group = df.groupby('lunch')[['math_score', 'reading_score', 'writing_score', 'total_score', 'average_score']].mean().reset_index()

# Melt the DataFrame for easier plotting
lunch_melted = lunch_group.melt(id_vars='lunch', var_name='Score Type', value_name='Score')

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=lunch_melted, x='Score Type', y='Score', hue='lunch', palette='Set2')

plt.title('Impact of Lunch Type on Student Performance (All Scores)', fontweight='bold')
plt.xlabel('Score Type')
plt.ylabel('Mean Score')
plt.legend(title='Lunch Type')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

### Insights
-Students who get Standard Lunch tend to perform better than students who got free/reduced lunch
### TEST PREPARATION COURSE COLUMN
- Which type of lunch is most common amoung students ?

### BIVARIATE ANALYSIS ( Is Test prepration course has any impact on student's performance ? )

```python
df['total_score'] = df[['math_score', 'reading_score', 'writing_score']].sum(axis=1)
df['average_score'] = df['total_score'] / 3

prep_group = df.groupby('test_preparation_course')[['math_score', 'reading_score', 'writing_score', 'total_score', 'average_score']].mean().reset_index()

prep_melted = prep_group.melt(id_vars='test_preparation_course', var_name='Score Type', value_name='Score')


plt.figure(figsize=(10, 6))
sns.barplot(data=prep_melted, x='Score Type', y='Score', hue='test_preparation_course', palette='Set2')

plt.title("Impact of Test Preparation Course on Student Performance", fontweight='bold')
plt.xlabel("Score Type")
plt.ylabel("Mean Score")
plt.legend(title='Test Preparation')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

### Insights
- Students who have completed the Test Prepration Course have scores higher in all the exams
